```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
file_path = '/Users/jhenriqueax/Desktop/Projeto IA/Dados/results_data.csv'
df = pd.read_csv(file_path)
```

```python
acerto = (df['label'] == df['classification']).mean() * 100
erro = (df['label'] != df['classification']).mean() * 100

print(f"A porcentagem de classificações acertadas: {acerto:.2f}%")
print(f"A porcentagem de classificações erradas: {erro:.2f}%")
```

```python
y_true = df['label']
y_pred = df['classification']

cm = confusion_matrix(y_true, y_pred)

disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap=plt.cm.Blues)
plt.show()
```
