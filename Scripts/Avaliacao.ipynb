# Treinamento de um Modelo para Classificação de Tweets com Discursos de Ódio

### Contém o código que verifica a acurária para nosso dataset de avaliação e monta a matriz de confusão

Membros: Lucas de Medeiros Soares, João Henrique Almeida Xavier e Tarso Jabbes Lima de Oliveira

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

## Import dos Dataset com os resultados

```python
file_path = '../Dados/results_data2.csv'
df = pd.read_csv(file_path)
```

## Avaliação de acertos e erros do modelo

```python
acerto = (df['label'] == df['classification']).mean() * 100
erro = (df['label'] != df['classification']).mean() * 100

print(f"A porcentagem de classificações acertadas: {acerto:.2f}%")
print(f"A porcentagem de classificações erradas: {erro:.2f}%")
```

## Construção da Matriz de Confusão

```python
y_true = df['label']
y_pred = df['classification']

cm = confusion_matrix(y_true, y_pred)

disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap=plt.cm.Blues)
plt.show()
```
