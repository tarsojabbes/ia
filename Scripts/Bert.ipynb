### Import das bibliotecas

```python
import pandas as pd
from transformers import BertTokenizer, BertForSequenceClassification
import torch
from sklearn.metrics import classification_report, accuracy_score
```

### Criação do dataset com a base de dados

```python
file_path = '/content/test_data.csv'
data = pd.read_csv(file_path)
```

### Tokenização dos textos

```python
tokenizer = BertTokenizer.from_pretrained('neuralmind/bert-base-portuguese-cased')
model = BertForSequenceClassification.from_pretrained('neuralmind/bert-base-portuguese-cased', num_labels=2)

inputs = tokenizer(data['query'].tolist(), padding=True, truncation=True, return_tensors="pt")
```

### Avaliação do acurácia com modelo sem Fine Tuning

```python
model.eval()
with torch.no_grad():
    outputs = model(**inputs)
    predictions = torch.argmax(outputs.logits, dim=-1)

accuracy = accuracy_score(data['label'], predictions)
classification_rep = classification_report(data['label'], predictions, target_names=["0", "1"])
```

```python
print("Accuracy:", f'{accuracy:.2f}')
print("Classification Report:\n", classification_rep)
```
